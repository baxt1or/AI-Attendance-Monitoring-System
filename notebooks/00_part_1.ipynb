```python
import pickle

DB_PATH = folder_path + 'embeddings.pkl'

with open(DB_PATH, "rb") as f:
    db = pickle.load(f)
```

```python
db.keys()
```

```python
def cosine_similarity(a, b):
    return np.dot(a, b) / (np.linalg.norm(a) * np.linalg.norm(b))
```

```python
sim = cosine_similarity(db["230035"], db["test3"])


if sim >= 0.6:
    print(f"Mathch with: {sim:2f}") 
else:
    print(f'No match: {sim:2f}')
```
